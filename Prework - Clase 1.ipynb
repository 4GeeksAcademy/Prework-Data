# Variables

Una variable es un contenedor para almacenar datos

```python
nombre = "Juan"  # Variable de tipo str
edad = 8         # Variable de tipo int
altura = 1.75     # Variable de tipo float
es_estudiante = True  # Variable de tipo bool
```

**convenciones de nomenclatura**: Se recomienda usar nombres descriptivos y usar minúsculas y separar palabras con guiones bajos (snake_case). Ejemplo: edad_usuario.

```python
print(nombre)
```

**Reasignar valores:** las variables en Python son mutables y se puede modificar el valor que se les asigna. Python siempre tendra en cuenta el ultimo valor asignado a una misma variable

```python
nombre = "Maria"
```

```python
print(nombre)
```

```python
Nombre = "Juan"
print(Nombre)
```

#Tipos de Datos

Representan el tipo de valor que una variable puede almacenar. Los más comunes son int, float, str, bool, listas, tuplas, y diccionarios.
Es importante esta distinción porque distintas funciones requieren distintos tipos de datos como argumentos.



```python
numero = 42        # int
precio = 19.99     # float
texto = "hola"     # str
es_valido = False  # bool
lista = [1, 2, 3]  # lista
tupla = (4, 5, 6)  # tupla
diccionario = {"clave": "valor"}  # diccionario
```

```python
print(len(texto))  # len() necesita una str, lista o tupla, no un int - Como saber que necesita cada funcion? -> buscar. recomendacion: https://www.w3schools.com/python/
```

¿Qué diferencia hay entre una lista y una tupla?

Las listas son mutables (se pueden cambiar después de crearse), mientras las tuplas son inmutables.

# Funciones

Una función es un bloque de código reutilizable que realiza una tarea específica. nos detendremos mas en las funciones en la siguiente clase.
conceptos claves:


*   se declaran con "def" antes del nombre de la funcion
*   siempre tiene que haber una salida de la funcion como "return" por ejemplo



```python
def saludar(nombre):
    return f"Hola, {nombre}!"

print(saludar(nombre))
```

# Estructuras de control - Condicionales

 Permiten ejecutar diferentes bloques de código dependiendo de si una condición es verdadera o falsa.

```python
if edad >= 18:
    print("Eres mayor de edad")
elif edad > 12:
    print("Eres adolescente")
else:
    print("Eres niño")
```

# Estructuras de control - Bucles e Iteración

permiten repetir bloques de código varias veces.  tip:  Usa for cuando sabes cuántas iteraciones necesitas y while cuando depende de una condición. Cuidados con while, si no incrementamos la variable, el bucle se se ejecutará indefinidamente, lo que puede bloquear el programa!!!

```python
# Bucle for
for a in range(5):
    print(a)
```

```python
# Bucle sobre una cadena de texto
for letra in "Python":
    print(letra)
```

```python
# Bucle while
x = 0
while x < 5:
    print(x)
    x += 1
```
